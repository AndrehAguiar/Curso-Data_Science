## Data

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

** Importe matplotlib.pyplot como plt e defina% matplotlib inline se você estiver usando o notebook jupyter. Qual comando você usa se você não estiver usando o notebook jupyter? **

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

## Exercício 1

** Acompanhe estes passos: **
* ** Crie um objeto de figura chamado fig usando plt.figure () **
* ** Use add_axes para adicionar um eixo à tela de figura em [0,0,1,1]. Chame esse novo eixo de "ax". **
* ** Plote (x, y) nesses eixos e defina os rótulos e títulos para corresponder ao gráfico abaixo: **

```python
def get_fig(position:list=[0,0,1,1], figsize:tuple=(6,4))->plt.axes:
    fig = plt.figure(figsize=figsize)
    return fig, fig.add_axes(position)

def set_plot_info(ax:plt.axes, title:str='TITLE', xlabel:str='x', ylabel:str='y')->plt.axes:
    return ax.set(title = title, xlabel = xlabel, ylabel = ylabel)
```

```python
_, ax = get_fig([0,0,1,1])
ax.plot(x,y)
set_plot_info(ax)
plt.show()
```

## Exercício 2
** Crie um objeto de figura e coloque dois eixos sobre ele, ax1 e ax2. Localizado em [0,0,1,1] e [0,2,0,5, .2, .2], respectivamente. **

```python
fig, ax1 = get_fig([0,0,1,1])
ax2 = fig.add_axes([.2,.5, .2, .2])
```

** Agora plote (x, y) em ambos os eixos. E chame seu objeto de figura para mostrá-lo. **

```python
ax1.plot(x,y, color='darkred')
set_plot_info(ax1)
ax2.plot(x,y, color='darkred')
set_plot_info(ax2, title='')
fig
```

## Exercício 3

** Crie o gráfico abaixo, adicionando dois eixos a um objeto de figura em [0,0,1,1] e [0,2,0,5, .4, .4] **

```python
fig, ax1 = get_fig([0,0,1,1])
ax2 = fig.add_axes([.2,.5, .4, .4])
```

** Agora use x, y e z arrays para recriar o gráfico abaixo. Observe os limites xlimits e y no gráfico inserido: **

```python
fig, ax1 = get_fig([0,0,1,1])
ax1.plot(x,z)
set_plot_info(ax1, title='X Vs X**2', ylabel='Z  (X**2)')

ax2 = fig.add_axes([.2,.5, .4, .4])
ax2.plot(x,z)
set_plot_info(ax2, title='ZOOM', ylabel='Z')
ax2.set_xlim(0,5)
ax2.set_ylim(0,25)
plt.show()
```

## Exercício 4

** Use plt.subplots (nrows = 1, ncols = 2) para criar o gráfico abaixo. **

```python
fig,ax = plt.subplots(1,2)
```

** Agora trace (x, y) e (x, z) nos eixos. Note a largura de linha e o estilo objetivos. **

```python
ax[0].plot(x,y, ls='--', lw=3, c='b',dash_joinstyle='bevel')
ax[1].plot(x,z, ls='-', lw=3, c='r')
fig
```

** Veja se você pode redimensionar o gráfico adicionando o argumento figsize () em plt.subplots () apenas copiando e colando seu código anterior. **

```python
_, ax = plt.subplots(1,2, figsize=(12,2))

ax[0].plot(x,y, ls='-', lw=3, c='b')
ax[0].tick_params(direction='in', length=6, width=2, colors='b',
               grid_color='b', grid_alpha=0.5)
set_plot_info(ax[0],title='x Vs y')

ax[1].plot(x,z, ls='--', lw=3, c='r')
ax[1].tick_params(direction='in', length=6, width=2, colors='r',
               grid_color='r', grid_alpha=0.5)
set_plot_info(ax[1],title='x Vs x**2', ylabel='z')
plt.show()
```
