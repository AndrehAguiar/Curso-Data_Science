# Exercícios

** Recrie os plots abaixo usando o DataFrame Titanic. Há poucas dicas, uma vez que a maioria dos plots pode ser feito com apenas uma ou duas linhas de código e uma dica basicamente daria a solução. Mantenha atenção aos rótulos x e y para dicas. **

** *Nota! Para não perder a imagem do plot, certifique-se de não codificar na célula que está diretamente acima do gráfico, há uma célula extra acima daquela que não substituirá esse gráfico! * **

```python
# PONHA O SEU CÓDIGO AQUI, 
# POIS SE USASSE A LINHA DEBAIXO, PODERIA ACABAR APAGANDO O PLOT QUE DESEJAMOS ENCONTRAR.
import scipy.stats as stat
ax = sns.jointplot('fare','age', titanic, s=20)
ax.ax_marg_x.set_xlim(-100, 600)
ax.ax_marg_y.set_ylim(-10,90)
ax.annotate(stat.pearsonr)
plt.show()
```

```python
# PONHA O SEU CÓDIGO AQUI, 
# POIS SE USASSE A LINHA DEBAIXO, PODERIA ACABAR APAGANDO O PLOT QUE DESEJAMOS ENCONTRAR.
ax = sns.distplot(titanic['fare'],
                  kde=False, bins=28, color='r',
                  hist_kws=dict(edgecolor="gray"))
ax.set_xlim(0, 600)
ax.set_ylim(0, 500)
plt.show()
```

```python
# PONHA O SEU CÓDIGO AQUI, 
# POIS SE USASSE A LINHA DEBAIXO, PODERIA ACABAR APAGANDO O PLOT QUE DESEJAMOS ENCONTRAR.
sns.boxplot('class','age', data=titanic,
            hue='class', dodge=False)

plt.show()
```

```python
# PONHA O SEU CÓDIGO AQUI, 
# POIS SE USASSE A LINHA DEBAIXO, PODERIA ACABAR APAGANDO O PLOT QUE DESEJAMOS ENCONTRAR.
sns.swarmplot('class', 'age', data=titanic,
              hue='class', alpha=0.7)
```

```python
titanic.groupby('sex')['who'].count()
```

```python
# PONHA O SEU CÓDIGO AQUI, 
# POIS SE USASSE A LINHA DEBAIXO, PODERIA ACABAR APAGANDO O PLOT QUE DESEJAMOS ENCONTRAR.
sns.countplot(x=titanic['sex'], hue=titanic['sex'], dodge=False)
plt.show()
```

```python
# PONHA O SEU CÓDIGO AQUI, 
# POIS SE USASSE A LINHA DEBAIXO, PODERIA ACABAR APAGANDO O PLOT QUE DESEJAMOS ENCONTRAR.
sns.heatmap(titanic.corr(), cmap="coolwarm")
```

```python
# PONHA O SEU CÓDIGO AQUI, 
# POIS SE USASSE A LINHA DEBAIXO, PODERIA ACABAR APAGANDO O PLOT QUE DESEJAMOS ENCONTRAR.
f = sns.FacetGrid(titanic, col='sex')
f.map(plt.hist, 'age')
```
