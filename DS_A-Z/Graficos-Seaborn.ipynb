## Trabalhando com Seaborn


```python
import seaborn as sb
sb.set_style('darkgrid')
```

```python
datasets = sb.get_dataset_names()
datasets
```

```python
titanic = sb.load_dataset('titanic')
titanic.head()
```

```python
tips = sb.load_dataset('tips')
tips.head()
```

```python
fly = sb.load_dataset('flights')
fly.head()
```

```python
fly = fly.pivot_table(values='passengers',index='month',columns='year')
fly
```

## Tipos de Gráficos
### Histograma

```python
tips.head()
```

```python
sb.distplot(tips['tip'], kde=False, bins=10)
```

```python
sb.distplot(tips['tip'], kde=False, bins=20)
```

## Correlacionando Gráficos
### Gráficos Lineares
## Joinplot
#### Junta histogramas com plots e é útil para ver a relação de duas variáveis

```python
sb.jointplot(x='total_bill', y='tip', data=tips)
```

```python
sb.jointplot(x='fare', y='age', data=titanic)
```

```python
sb.jointplot(x='size', y='tip', data=tips)
```

### Analisando os dados de maneira geral com Pairplot

```python
sb.pairplot(tips)
```

```python
tips.time.isnull().sum()
```

```python
sb.pairplot(tips, hue='time', diag_kind='hist',
            diag_kws=dict(alpha=0.5))
```

```python
sb.pairplot(tips, hue='smoker', diag_kind='hist',
            diag_kws=dict(alpha=0.5))
```

```python
sb.pairplot(tips, hue='sex', diag_kind='hist', palette='PRGn',
            diag_kws=dict(alpha=0.9))
```

## Plotando dados categóricos

```python
sb.countplot(y='fare', data=titanic)
```

```python
sb.countplot(x='sex', data=titanic)
```

## Boxplot

```python
sb.boxplot(x='day', y='tip', data=tips)
```

```python
sb.boxplot(x='day', y='tip', data=tips, hue='sex')
```

```python
sb.violinplot(x='day', y='tip', data=tips)
```

```python
sb.violinplot(x='day', y='tip', data=tips, hue='sex')
```

```python
sb.violinplot(x='day', y='tip', data=tips, hue='sex', split=True)
```

## Stripplot

```python
sb.stripplot(x='day', y='tip', data=tips, jitter=False)
```

```python
sb.stripplot(x='day', y='tip', data=tips, jitter=True)
```

```python
sb.swarmplot(x='day', y='tip', data=tips)
```

```python
sb.swarmplot(x='day', y='tip', data=tips, color='k')
sb.violinplot(x='day', y='tip', data=tips)
```
