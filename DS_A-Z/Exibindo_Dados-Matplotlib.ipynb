```python
def calcula_escala(p_medio, lst_size):
    const = p_medio/(lst_size//2)
    i = (lst_size//2)*(-1)
    j = i*(-1)
    ticks = []
    while i <= j:
        t = p_medio+(const*i) if i != 0 else p_medio
        ticks.append(t)
        i+=1
    return ticks
```
