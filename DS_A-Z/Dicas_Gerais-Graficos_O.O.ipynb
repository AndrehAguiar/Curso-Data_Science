```python
import matplotlib.pyplot as plt
import random as rd
%matplotlib inline
```

```python
dias = []
func01 = []
func02 = []

for i in range(1,366):
    dias.append(i)
    
    x = rd.randint(0,100)
    func01.append(x)
    
    y = rd.randint(0,100)
    func02.append(y)
    
print(dias)
print(func01)
print(func02)
```

```python
plt.plot(dias, func01, c='r', ls=':')
plt.plot(dias, func02, c='b',ls=':')
```

```python
meses = []
func01 = []
func02 = []
func03 = []

for i in range(1,31):
    meses.append(i)
    
    x = rd.randint(0,100)
    func01.append(x)
    
    y = rd.randint(0,100)
    func02.append(y)
    
    z = rd.randint(25,70)
    func03.append(z)
    
print(dias)
print(func01)
print(func02)
print(func03)
```

```python
plt.text(11,93, 'Texto no grfico', color='k',
         style='oblique', weight='heavy',
         bbox=dict(facecolor='yellow',alpha=0.3))

plt.plot(meses, func01, c='r', ls=':')
plt.plot(meses, func02, c='b',ls=':')
plt.plot(meses, func03, c='g')
```

```python
plt.annotate(xy=(10,20), xytext=(12,90), s='Anotao',
             color='k', style='oblique', weight='heavy',
             bbox=dict(facecolor='yellow',alpha=0.3),
             arrowprops = dict(arrowstyle = '-'))

plt.plot(meses, func01, c='r', ls=':')
plt.plot(meses, func02, c='b',ls=':')
plt.plot(meses, func03, c='g')
```

## Grficos orientados a Objetos

```python
plt.figure(figsize=(10,5), dpi=150)
plt.subplot(1,3,1)
plt.plot(meses, func01, c='r', ls=':')

plt.subplot(1,3,2)
plt.plot(meses, func02, c='b', ls=':')

plt.subplot(1,3,3)
plt.plot(meses, func03, c='g', ls='-')

plt.tight_layout()
```

```python
x = list(range(1,13))
y = [34,65,32,65,67,89,35,65,45,89,56,45]

figura = plt.figure()

axes01 = figura.add_axes([0.1,0.1,0.8,0.8])
axes01.plot(x,y)
axes01.set_xlabel('Meses')
axes01.set_ylabel('Produo')
```

```python
meses = list(range(1,13))
func01 = [34,65,32,65,67,89,35,65,45,89,56,45]
func02 = [34,36,32,32,60,23,30,35,26,28,24,30]

figura = plt.figure()

axes01 = figura.add_axes([0.1,0.1,0.8,0.8])
axes02 = figura.add_axes([0.2,0.5,0.3,0.3])

axes01.plot(meses,func01, c='r')
axes02.plot(meses, func02)

axes01.set_xlabel('Meses')
axes01.set_ylabel('Produo')
```

```python
fig, ax = plt.subplots()

ax.plot(meses, func01)
ax.plot(meses, func02, c='r')
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 2)

plt.tight_layout()

ax[0].plot(meses, func01)
ax[1].plot(meses, func02)
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize=(10,6), dpi=150)

plt.tight_layout()

ax[0].plot(meses, func01, label = 'Funcionrio 01')
ax[1].plot(meses, func02, label = 'Funcionrio 02')
```
