```python
def break_string(text:str)->str:
    """
    Recebe uma string e identifica todas as palavras iniciadas com letra maiúscula
    que não são antecedidas por espaços e retorna cada palavra separada por ', '
    """
    lst_idx = []
    string = ''
    
    for idx, char in enumerate(text[1:]):
        if char.isupper() and (text[1:][idx-1] != ' ') and (text[1:][idx-1] != '-'):
            lst_idx.append(idx)
            
    if len(lst_idx) > 0:
        for i, idx in enumerate(lst_idx):
            string = f'{string}, {text[lst_idx[i-1]+1:idx+1]}' if i > 0 else text[:idx+1]

        return f'{string}, {text[lst_idx[-1]+1:]}'
    else:
        return text
            
```

```python
drinks_6.sum()['country'].apply(lambda x: break_string(x))
```
